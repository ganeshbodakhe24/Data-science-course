```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,40,50,20,40,60,100,30,40,30,50]
x1=[10,40,50,60,20,30,70,30,20,50,30,20]
```

```python
plt.figure(figsize=(4,4))
plt.boxplot(x)
plt.show()
```

```python
# two box plot
y=[x1,x]
plt.boxplot(y,labels=["py","c"])
plt.show()
```

```python
# widths
plt.figure(figsize=(4,4))
plt.boxplot(x,widths=0.6)
plt.show()
```

```python

plt.figure(figsize=(4,4))
plt.boxplot(x,widths=0.6,labels=["py"])
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(4,4))
plt.boxplot(x,widths=0.1,labels=["py"],notch=True)
plt.legend()
plt.show()
```

```python
# outliers symbole change

plt.figure(figsize=(4,4))
plt.boxplot(x,widths=0.1,labels=["py"],notch=True,sym="b+",showcaps=bool)
plt.show()
```

```python
# patch_artist='red' -to fill the box plot box
plt.figure(figsize=(4,4))
plt.boxplot(x,widths=0.1,labels=["py"],notch=True,sym="b+",patch_artist=True,showcaps=bool)
plt.show()
```

```python
# outliers symbole change

plt.figure(figsize=(4,4))
plt.boxplot(x,widths=0.1,labels=["py"],notch=True,sym="b+",showcaps=bool,showmeans=True,showbox=True)
plt.show()
```

```python
# outliers symbole change

plt.figure(figsize=(4,4))
plt.boxplot(x,widths=0.1,labels=["py"], boxprops=dict(color="r"),medianprops=dict(color="b") )
plt.show()
```
