```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[1,3,5,7,9,5,3,2,3]
```

```python
plt.figure(figsize=(4,3))
plt.stem(x,y)
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.stem(x,y, markerfmt="r+")
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.stem(x,y, markerfmt="r+", basefmt="g",label="python",bottom=5)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.stem(x,y, markerfmt="r+",linefmt=":", basefmt="g",label="python",bottom=5,orientation='horizontal')
plt.legend()
plt.show() 
```
