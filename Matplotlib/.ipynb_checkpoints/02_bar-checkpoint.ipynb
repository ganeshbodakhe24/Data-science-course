```python
import matplotlib.pyplot as plt
```

```python
x=["python",'java',"c","c++"]
y=[20,30,40,50]
```

```python
plt.figure(figsize=(3,2))
plt.bar(x,y)
```

# color

```python

plt.figure(figsize=(3,2))
plt.bar(x,y,color="r")
plt.show()
```

```python
plt.figure(figsize=(3,2))
c=["red","green","yellow","pink"]
plt.bar(x,y,color=c)
plt.show()
```

# label

```python
plt.figure(figsize=(3,2))
plt.bar(x,y)
plt.xlabel("Languages")
plt.ylabel("Performance")
plt.show()
```

```python
plt.figure(figsize=(3,2))
plt.bar(x,y)
plt.xlabel("Languages",fontsize=30)
plt.ylabel("Performance" ,fontsize=30)
plt.title("My Languare Progress")
plt.show()
```

```python
plt.figure(figsize=(3,2))
plt.bar(x,y,width=0.2)
plt.xlabel("Languages",)
plt.ylabel("Performance" ,)
plt.show()
```

```python
plt.figure(figsize=(3,2))
plt.bar(x,y,width=0.7,align="edge")
plt.show()
```

# edge color

```python
plt.figure(figsize=(3,2))
plt.bar(x,y,width=0.7,edgecolor="r")
plt.show()
```

# width of edge

```python
plt.figure(figsize=(3,2))
plt.bar(x,y,width=0.7,edgecolor="red",linewidth=5)
plt.show()
```

```python
plt.figure(figsize=(3,2))
plt.bar(x,y,width=0.7,edgecolor="red",linewidth=5,linestyle=":")
plt.show()
```

# legend

```python
plt.figure(figsize=(5,3))
plt.bar(x,y)
plt.xlabel("Languages")
plt.ylabel("Performance")
plt.legend("hii")
plt.show()
```

```python
plt.figure(figsize=(5,3))
x=["java","c","c++","python"]
y=[60,70,80,90]
z=[20,30,40,50]
plt.bar(x,y,color="y")
plt.bar(x,z,color="r")
```

```python
plt.figure(figsize=(5,3))
x=["java","c","c++","python"]
y=[60,70,80,90]
z=[20,30,40,50]
plt.bar(x,y,color="y",label="ganesh")
plt.bar(x,z,color="r",label="sham")
plt.legend()
plt.show()
# but the are overlapping
```

```python
import numpy as np
plt.figure(figsize=(5,3))
x=["java","c","c++","python"]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
y=[60,70,80,90]
z=[20,30,40,50]
plt.bar(p,y,color="y",width=width,label="ganesh")
plt.bar(p1,z,color="r",width=width,label="sham")
plt.legend()
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(5,3))
x=["java","c","c++","python"]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
y=[60,70,80,90]
z=[20,30,40,50]
plt.xticks(p+width/2,x,rotation=45)
plt.bar(p,y,color="y",width=width,label="ganesh")
plt.bar(p1,z,color="r",width=width,label="sham")
plt.legend()
plt.show()
```

# horizontal graph

```python
plt.figure(figsize=(3,2))
plt.barh(x,y,height=0.4)
plt.xlabel("Languages",)
plt.ylabel("Performance" ,)
plt.show()
```
