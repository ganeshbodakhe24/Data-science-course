# Histogram

```python
import matplotlib.pyplot as plt
```

```python
x=[10,50,30,20,70,60,50,25,60,30,60]
```

```python
plt.figure(figsize=(4,3))
plt.hist(x)
plt.show()
```

# labels

```python
plt.figure(figsize=(4,3))
plt.hist(x)
plt.title("this is hist",fontsize=30)
plt.xlabel("marks")
plt.ylabel("repeate")
plt.show()
```

```python
# botom label starting position 
plt.figure(figsize=(4,3))
plt.hist(x,align='right')
# right ,mid,left
plt.show()
```

```python
# virtical numbering starting
plt.figure(figsize=(4,3))
plt.hist(x,bottom=20)
plt.show()
```

```python
# legend label
plt.figure(figsize=(4,3))
plt.hist(x,bottom=20,label="hii")
plt.legend()
plt.show()
```

# color

```python

plt.figure(figsize=(4,3))
plt.hist(x,bottom=20,label="hii",color="red")
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(4,3))
col=["red","green","yellow","pink","green","yellow","pink"]
# //we cant apply more than one color

plt.hist(x,bottom=20,label="hii",color='green')
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(4,3))
plt.hist(x,bottom=20,edgecolor="green",linewidth=10)
plt.show()
```

# rwidth-bar width

```python

plt.figure(figsize=(4,3))
plt.hist(x,bottom=20,rwidth=0.4)
plt.show()
```

# type of histogram

```python

plt.figure(figsize=(4,3))
plt.hist(x,bottom=20,histtype='bar')
# step,stepfiled,bar,barfiled
plt.show()
```
