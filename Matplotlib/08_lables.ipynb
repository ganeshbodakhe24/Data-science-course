# stack plot or area

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8]
a1=[2,5,2,6,8,5,3,1]
a2=[4,5,6,7,3,2,4,5]
a3=[1,2,4,5,6,7,4,6]
```

```python
la=["area1","area2","area3"]
plt.stackplot(x,a1,a2,a3,labels=la)
plt.legend()
plt.show()
```

```python
la=["area1","area2","area3"]
plt.stackplot(x,a1,a2,a3,labels=la)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("area")
plt.show()
```

```python
la=["area1","area2","area3"]
plt.stackplot(x,a1,a2,a3,labels=la,colors=["r","g","y"])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("area")
plt.show()
```

```python
la=["area1","area2","area3"]
plt.stackplot(x,a1,a2,a3,labels=la,baseline='weighted_wiggle')
# baseline{'zero', 'sym', 'wiggle', 'weighted_wiggle'}
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("area")
plt.show()
```

```python

la=["area1","area2","area3"]
plt.stackplot(x,a1,a2,a3,labels=la)
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("area")
plt.show()
```
