# fill between

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]
y=[2,3,4,5,6,7,8]
```

```python
plt.figure(figsize=(2,2))
plt.plot(x,y)
plt.fill_between(x,y)
plt.show()
```

```python
plt.figure(figsize=(2,2))
plt.plot(x,y)
plt.fill_between(x,y,color="red")
plt.show()
```

```python
# plt.figure(figsize=(2,2))
plt.plot(x,y)
plt.fill_between(x,y,color="red",alpha=0.1,step="mid")
# step{'pre', 'post', 'mid'}, optional
plt.show()
```

```python
import numpy as np
x=np.array([1,2,3,4,5,6,7])
y=np.array([2,3,4,5,6,7,8])
```

```python
# plt.figure(figsize=(2,2))
plt.plot(x,y)
plt.fill_between(x,y,color="red",alpha=0.1,where=(x>=2) & (x<4))
# step{'pre', 'post', 'mid'}, optional
plt.show()
```
