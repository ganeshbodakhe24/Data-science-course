```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]
y=[1,2,3,4,5,6,7]
```

```python
plt.plot(x,y)
# xticks is helpful to get x axis number according to our input ddata 
# xlim and ylim is use to set x axis limit of x
plt.xlim(0,10)
plt.ylim(0,10)
plt.xticks(x)
plt.yticks(y)
plt.show()
```
