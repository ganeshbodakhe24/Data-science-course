# pie plot

```python
from matplotlib import pyplot as plt
```

```python
x=[10,40,30,20]
y=['java','python','c','HTML']
```

```python
plt.figure(figsize=(4,3))
plt.pie(x,labels=y)
plt.show()
```

autopct-None or str or callable, default: None
If not None, autopct is a string or function used to label the wedges with their numeric value.
The label will be placed inside the wedge. If autopct is a format string, the label will be fmt % pct. 
If autopct is a function, then it will be called.

```python
# percentage
plt.figure(figsize=(4,3))
plt.pie(x,labels=y,autopct='%0.1f%%')
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.pie(x,labels=y,autopct='%0.1f%%')
plt.show()
```

```python
# colors
col=["red","green","pink","yellow"]
plt.figure(figsize=(4,3))
plt.pie(x,labels=y,autopct='%0.1f%%',colors=col)
plt.show()
```

```python
# explode 
ex=[0.2,0.0,0.0,0.0]
plt.figure(figsize=(4,3))
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%')
plt.show()
```

```python
#  
ex=[0.2,0.0,0.0,0.0]
plt.figure(figsize=(4,3))
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%',shadow=True,labeldistance=1.1)
plt.show()
```

```python
x1=[2,4,3,6,5,3] 
plt.figure(figsize=(4,3))
plt.pie(x1,autopct='%0.1f%%',shadow=True,labeldistance=1.1)
plt.show()
```

```python
# start angle

ex=[0.2,0.0,0.0,0.0]
plt.figure(figsize=(4,3))
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%',shadow=True,labeldistance=1.1,startangle=40,radius=0.9)
plt.show()
```

```python
# counter clockwise

ex=[0.2,0.0,0.0,0.0]
plt.figure(figsize=(4,3))
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%',shadow=True,labeldistance=1.,counterclock=False,rotatelabels=True,pctdistance=0.5,wedgeprops = {'linewidth': 3})
plt.show()
```

```python
x1=[1]
# plt.pie(x1)
plt.pie([1])
plt.show()
```

```python
# counter clockwise

ex=[0.2,0.0,0.0,0.0]
plt.figure(figsize=(4,3))

plt.pie(x,labels=y)
plt.pie([1],radius=0.8,colors='w')
plt.legend()
plt.show()
```
