```python
# distribution plot
sns.distplot(dataset["CoapplicantIncome"])
```

```python
# now find third normal distribution min_value and max_value
# min_value=mean-3*std
min_value=dataset["CoapplicantIncome"].mean()-(3*dataset["CoapplicantIncome"].std())
min_value
```

```python
# now find third normal distribution min_value and max_value
# max_value=mean+3*std
max_value=dataset["CoapplicantIncome"].mean()+(3*dataset["CoapplicantIncome"].std())
max_value
```

```python
# now our value is in negative form hance there is no negative outlier
```

```python
# positive outliter 
new_dataset=dataset[dataset["CoapplicantIncome"]<max_value]
```
