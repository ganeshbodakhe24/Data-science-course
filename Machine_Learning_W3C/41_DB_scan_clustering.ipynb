```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=250,noise=0.06)
```

```python
x
```

```python
data={"data1":x[:,0],"data2":x[:,1]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
sns.scatterplot(x="data1",y='data2',data=df)
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbs=DBSCAN(eps=0.3,min_samples=4)
```

```python
df["predict"]=pd.DataFrame(dbs.fit_predict(df))
```

```python
df
```

```python
sns.scatterplot(x="data1",y='data2',data=df, hue="predict")
```
