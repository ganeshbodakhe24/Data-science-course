```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("loan.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
ss.fit(dataset[["CoapplicantIncome"]])
```

```python
arr=ss.transform(dataset[["CoapplicantIncome"]])
arr
```

```python
dataset["CoapplicantIncome_ss"]=pd.DataFrame(arr,columns=["x"])
```

```python
dataset
```

```python
# before data
plt.figure(figsize=(20, 7))
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(dataset["CoapplicantIncome"])
plt.subplot(1,2,2)
plt.title("after")
sns.distplot(dataset["CoapplicantIncome_ss"])

```
