```python
# we can check the accuracy of model by cross validation  
# for that we split the data input  into  different type of train test data
# k-fold cross validation 
#leaveOneOut cross validation
#leavePOut cross validation 
#stratified k fold cross validation - will work on classified data 
```

```python
import pandas as pd 
```

```python
df=pd.read_csv("placement.csv")
```

```python
df
```

```python
new_df=df.head(10)
```

```python
new_df
```

```python
# //only 10 record taken for analysis purpose
```

```python
x=new_df[['cgpa']]
y=new_df["package"]
```

```python
from sklearn.model_selection import LeaveOneOut,LeavePOut,KFold,StratifiedKFold
```

```python
lo=LeaveOneOut()
for train,test in lo.split(x,y):
    print(train,test)
```

```python
lo=LeavePOut(p=2)
for train,test in lo.split(x,y):
    print(train,test)
```

```python
lo=KFold(n_splits=4)
for train,test in lo.split(x,y):
    print(train,test)
```
