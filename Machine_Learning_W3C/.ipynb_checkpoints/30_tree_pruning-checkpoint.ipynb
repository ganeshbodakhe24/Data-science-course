```python
# pruning 
# to minimizing the depth according to the accuracy of the model
# to stop model from from overfitting and underfitting we use pruning 
```

```python
#we simply add how much depth we want in model
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeClassifier
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("Social_Network_Ads.csv")
```

```python
df
```

```python
x=df.iloc[:,2:-1]
```

```python
y=df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.24)
```

```python
dtc=DecisionTreeClassifier(max_depth=5)
# to giveing max_depth firsly called as pre pruning
```

```python
dtc.fit(x_train,y_train)
```

```python
dtc.score(x_train,y_train)
```

```python
dtc.score(x_test,y_test)
```

```python
plt.figure(figsize=(20,20))
plot_tree(dtc)
```

```python
# post pruning
```

```python
# make loop and find out the difference in train and test score 
# which one have less difference in train and test score pass this to model again
```

```python
for i in range (1,20):
    dtc=DecisionTreeClassifier(max_depth=i)
    dtc.fit(x_train,y_train)
    print(dtc.score(x_train,y_train)," ", dtc.score(x_test,y_test) ,"- " ,i )
```

```python
# accorging to loop the depth 2 and 3 is better choice for make accurate tree
```
