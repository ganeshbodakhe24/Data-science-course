```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df=pd.read_csv('Iris.csv')
```

```python
df.drop(columns=["Id","Species"] ,inplace=True)
```

```python
df
```

```python
# we does not split the data into train and  test in unsupervise machine learning
```

```python
# we have to find the wcss value by that we can find out required number of cluster by elblow method
```

```python
from sklearn.cluster import KMeans
```

```python
# with the help of loop we increase value of k and find the wcss  value
# km.inertia_ is the value of wcss
```

```python
wcss=[]
for i in range(1,20):
    km=KMeans(n_clusters=i,init='k-means++')
    km.fit(df)
    wcss.append(km.inertia_) 
```

```python
wcss
```

```python
# now plot the elbow cruve
plt.plot([i for i in range(1,20)],wcss,marker="o")
plt.xlabel("no of cluster")
plt.ylabel('wcss')
plt.xticks([i for i in range(1,20)])
plt.grid(axis='x')
plt.show()
```

```python
 km=KMeans(n_clusters=3,init='k-means++')
 km.fit_predict(df)
```

```python
df["predicted"]=pd.DataFrame( km.fit_predict(df))
```

```python
km.labels_
```

```python
sns.pairplot(data=df,hue="predicted")
plt.show()
```
