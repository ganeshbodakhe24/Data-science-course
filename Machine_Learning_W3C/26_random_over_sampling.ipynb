```python
# random over sampling is use to fit the model properly to unbalance dataset
# like we have 100 sample for yes and 400 sample NO  output then this model is bias for no
 # to remove this we have to make ti equal in over sampling 
# both output no and yes is 400,400
```

```python
import pandas as pd
df=pd.read_csv("Social_Network_Ads.csv")
```

```python
df.drop(columns=['User ID','Gender'],inplace=True)
```

```python
df
```

```python
df['Purchased'].value_counts()
# data is bias on no 
```

```python
# to remove this use random over fitting
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
os=RandomOverSampler()
```

```python
x=df.iloc[:,:-1]
```

```python
y=df.iloc[:,-1]
```

```python
os_x,os_y=os.fit_resample(x,y)
```

```python
os_x
```

```python
os_y.value_counts()
# over sampled data 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(os_x,os_y,test_size=0.24,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```
