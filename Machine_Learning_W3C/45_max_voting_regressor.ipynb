```python
import pandas as pd
```

```python
df=pd.read_csv("placement.csv")
```

```python
df.head(3)
```

```python
x=df.iloc[:,:-1]
```

```python
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=32,test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
print(lr.score(x_train,y_train),": --:",lr.score(x_test,y_test))
```

```python
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
print(dt.score(x_train,y_train),": --:",dt.score(x_test,y_test))
```

```python
sv=SVR()
sv.fit(x_train,y_train)
print(sv.score(x_train,y_train),": --:",sv.score(x_test,y_test))
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
data=[("lr1",LinearRegression()),("sv1",SVR()),("dt",DecisionTreeRegressor())]
```

```python
vr=VotingRegressor(data)
```

```python
vr.fit(x_train,y_train)
```

```python
vr.score(x_test,y_test)
```

```python
demo={"lr":lr.predict(x_test),"sv":sv.predict(x_test),"dt":dt.predict(x_test),"vr":vr.predict(x_test)}
```
